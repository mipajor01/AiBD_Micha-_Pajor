```python
import numpy as np
import matplotlib.pyplot as plt

def kwardz_fun(x,min_a,max_a):
    y=[]
    for i in x:
        y.append(i**2+5)
    plt.plot(x,y,label='f^2+5')
    plt.grid('on')
    plt.title('Wykres funkcji dla przedzia≈Çu od' + str(min_a) + ' do ' + str(max_a))
    plt.xlabel('o≈õ x')
    plt.ylabel('o≈õ y')
    plt.xlim(min_a,max_a)
    plt.legend(loc = 4)
    plt.show()
x = np.linspace(-1,1,2000)
kwardz_fun(x,-1,1)
x = np.linspace(-6,6,2000)
kwardz_fun(x,-6,6)
x = np.linspace(0,5,2000)
kwardz_fun(x,0,5)
```
